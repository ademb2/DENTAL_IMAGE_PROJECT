```python
import pandas as pd
```

```python
df = pd.read_csv('/dental_image_project/data/raw/dental_images.csv')
df['category'] = df['category'].replace('Environment_cohort', 'Control_cohort')
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# Aggregate counts
df_counts = df.groupby(['image_type', 'category']).size().unstack(fill_value=0)

# Plotting
fig, ax = plt.subplots(figsize=(10, 6))

# Define colors for each category
colors = plt.get_cmap('tab20')

# Plot stacked bars
bottoms = np.zeros(len(df_counts))
for i, category in enumerate(df_counts.columns):
    ax.bar(df_counts.index, df_counts[category], label=category, color=colors(i), bottom=bottoms)
    bottoms += df_counts[category]

# Add labels and title
ax.set_xlabel('Image Type')
ax.set_ylabel('Count')
ax.set_title('Stacked Bar Chart with Categories')
ax.legend(title='Category')

# Show the plot
plt.tight_layout()
plt.show()
```
